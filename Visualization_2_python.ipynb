```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv("/content/tips project visualization.csv")
```

```python
data.head()
```

```python
#Violin plot
snr.catplot(x="total_bill",y="day",hue="sex",kind="violin",data=data)
```

```python
#Box Plot
snr.catplot(x="tip",y="day",hue="sex",kind="box",data=data)
```
